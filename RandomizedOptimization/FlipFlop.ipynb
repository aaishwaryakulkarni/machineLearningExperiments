```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import mlrose_hiive as mlrose
import matplotlib.pyplot as plt
```

```python
length_range = range(10,220,20)
```

```python
sa_dict = {}
rhc_dict = {}
ga_dict = {}
minic_dict = {}
```

```python
def flip_flop_length(len):
    
    fitness = mlrose.FlipFlop()
    problem = mlrose.DiscreteOpt(length = len,
                                 fitness_fn = fitness,
                                 maximize = True,
                                 max_val = 2)
    problem.set_mimic_fast_mode(True)
    
    #--------------------------------Simulated Annealing------------------------------------------------
    _,sa_fitness,_ = mlrose.simulated_annealing(problem,
                                                schedule = mlrose.ExpDecay(),
                                                max_attempts = 100,
                                                max_iters = 1000,
                                                init_state = np.random.randint(2, size = len),
                                                curve = True)
    sa_dict[len] = sa_fitness

    #----------------------------------Randomized Hill Climbing------------------------------------------------

    _,rhc_fitness, _ = mlrose.random_hill_climb(problem,
                                                max_attempts = 100,
                                                max_iters = 1000,
                                                init_state = np.random.randint(2, size = len),
                                                curve = True)
    rhc_dict[len] = rhc_fitness
    
    #-------------------------------------Genetic Algorithm------------------------------------------------

    _,ga_fitness, _ = mlrose.genetic_alg(problem,
                                         max_attempts = 100,
                                         max_iters = 1000,
                                         curve = True)
    ga_dict[len] = ga_fitness
    
    #----------------------------------------MIMIC------------------------------------------------

    _,mimic_fitness, _ = mlrose.mimic(problem,
                                      max_attempts = 100,
                                      max_iters = 1000,
                                      curve = True)
    minic_dict[len] = mimic_fitness

def plot_lenght_graphs(plt,sa_dict,rhc_dict,ga_dict,minic_dict):
    plt.plot(rhc_dict.keys(),
             rhc_dict.values(),
             label = 'Randomized Hill Climb')
    plt.plot(sa_dict.keys(),
             sa_dict.values(),
             label = 'Simulated Annealing')
    plt.plot(ga_dict.keys(),
             ga_dict.values(),
             label = 'Genetic Algorithm')
    plt.plot(minic_dict.keys(),
             minic_dict.values(),
             label = 'MIMIC')
    plt.title('Fitness VS Problem Size (Flip Flop)')
    plt.xlabel('Problem Size')
    plt.ylabel('Fitness')
    plt.legend()
    
def flip_flop_iterations(problem_length):
    
    fitness = mlrose.FlipFlop()
    
    problem = mlrose.DiscreteOpt(length = problem_length,
                                       fitness_fn = fitness,
                                       maximize = True,
                                       max_val = 2)
    problem.set_mimic_fast_mode(True)
    init_state = np.random.randint(2, size = problem_length)
    
    _, _, rhc_fitness_curve = mlrose.random_hill_climb(problem,
                                                       max_attempts = 10,
                                                       max_iters = 2000,
                                                       init_state = init_state,
                                                       curve = True)
    _, _, sa_fitness_curve = mlrose.simulated_annealing(problem,
                                                        schedule = mlrose.ExpDecay(),
                                                        max_attempts = 10,
                                                        max_iters = 2000,
                                                        init_state = init_state,
                                                        curve = True)

    _, _, ga_fitness_curve = mlrose.genetic_alg(problem,
                                                max_attempts = 10,
                                                max_iters = 2000,
                                                curve = True)

    _, _, mimic_fitness_curve = mlrose.mimic(problem,
                                             max_attempts = 10,
                                             max_iters = 2000,
                                             curve = True)
    
    return rhc_fitness_curve,sa_fitness_curve,ga_fitness_curve,mimic_fitness_curve

def rhc_optimization(problem_length,restart_list):
    li =[]
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = problem_length,
                                   fitness_fn = fitness,
                                   maximize = True,
                                   max_val = 2)
    for restart_value in restart_list:
        _, _, rhc_fitness_curve = mlrose.random_hill_climb(problem,
                                                           restarts = restart_value,
                                                           max_attempts = 10,
                                                           max_iters = 2000,
                                                           init_state = np.random.randint(2, size = problem_length),
                                                           curve = True)
        li.append(rhc_fitness_curve)
    
    return li

def sa_optimization(problem_length,schedule_list):
    li = []
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = problem_length,
                                   fitness_fn = fitness,
                                   maximize = True,
                                   max_val = 2)
    
    for schedule in schedule_list:
    
        _, _, sa_fitness_curve = mlrose.simulated_annealing(problem,
                                                            schedule = schedule,
                                                            max_attempts = 10,
                                                            max_iters = 2000,
                                                            init_state = np.random.randint(2, size = problem_length),
                                                            curve = True)
        li.append(rhc_fitness_curve)
    
    return li

def ga_optimization(problem_length,hyperparam_list):
    li = []
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = problem_length,
                                   fitness_fn = fitness,
                                   maximize = True,
                                   max_val = 2)
    for i in hyperparam_list:
        print(i)
    
        _, _, ga_fitness_curve = mlrose.genetic_alg(problem,
                                                    mutation_prob = i[0],
                                                    pop_size = i[1],
                                                    max_attempts = 10,
                                                    curve = True)
        li.append(ga_fitness_curve)
    
    return li

def mimic_optimization(problem_length,hyperparam_list):
    li = []
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = problem_length,
                                 fitness_fn = fitness,
                                 maximize = True,
                                 max_val = 2)
    
    for i in hyperparam_list:
        print(i)
        
        _, _, mimic_fitness_curve = mlrose.mimic(problem,
                                                 keep_pct = i[0],
                                                 max_iters = 1000,
                                                 pop_size = i[1],
                                                 max_attempts = 10,
                                                 curve = True)
        li.append(mimic_fitness_curve)
    
    return li
```

```python
def final_optimized():
    
    time_dict = {}

    #--------------------------------Simulated Annealing------------------------------------------------
        
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = 200,
                                 fitness_fn = fitness,
                                 maximize = True,
                                 max_val = 2)
    problem.set_mimic_fast_mode(True)
    
    start = time.time()
    _,_,sa_curve= mlrose.simulated_annealing(problem,
                                             schedule = mlrose.ArithDecay(),
                                             max_attempts = 100,
                                             max_iters = 1000,
                                             init_state = np.random.randint(2, size = 200),
                                             curve = True)
    end = time.time()
    
    time_dict['SA'] = end - start

    #----------------------------------Randomized Hill Climbing------------------------------------------------
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = 200,
                                   fitness_fn = fitness,
                                   maximize = True,
                                   max_val = 2)
    problem.set_mimic_fast_mode(True)
    
    start = time.time()
    
    _,_,rhc_curve= mlrose.random_hill_climb(problem,
                                            restarts = 4,
                                            max_attempts = 100,
                                            max_iters = 1000,
                                            init_state = np.random.randint(2, size = 200),
                                            curve = True)
    end = time.time()
    
    time_dict['RHC'] = end - start
    #-------------------------------------Genetic Algorithm------------------------------------------------
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = 200,
                                 fitness_fn = fitness,
                                 maximize = True,
                                 max_val = 2)
    problem.set_mimic_fast_mode(True)
    
    start = time.time()
    _,_,ga_curve = mlrose.genetic_alg(problem,
                                      mutation_prob =0.1,
                                      pop_size = 150,
                                      max_attempts = 10,
                                      max_iters = 1000,
                                      curve = True)

    end = time.time()
    
    time_dict['GA'] = end - start
    #----------------------------------------MIMIC------------------------------------------------
    fitness = mlrose.FlipFlop()

    problem = mlrose.DiscreteOpt(length = 200,
                                   fitness_fn = fitness,
                                   maximize = True,
                                   max_val = 2)
    problem.set_mimic_fast_mode(True)
    
    start = time.time()
    
    _,_,mimic_curve = mlrose.mimic(problem,
                                   keep_pct = 0.4,
                                   pop_size = 150,
                                   max_attempts = 100,
                                   max_iters = 1000,
                                   curve = True)

    end = time.time()

    time_dict['MIMIC'] = end - start
    return sa_curve,rhc_curve,ga_curve,mimic_curve,time_dict
```

```python
sa_final,rhc_final,ga_final,mimic_final,time_dict = final_optimized()
```

```python
plt.figure()
plt.plot(rhc_final[:,0],
         label = 'Randomized Hill Climb')
plt.plot(sa_final[:,0],
         label = 'Simulated Annealing')
plt.plot(ga_final[:,0],
         label = 'Genetic Algorithm')
plt.plot(mimic_final[:,0],
         label = 'MIMIC')
plt.title('Fitness Curve (Flip Flop) Final Evaluation')
plt.legend()
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.show()
```
